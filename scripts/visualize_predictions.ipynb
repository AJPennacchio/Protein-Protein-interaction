```python
import pandas as pd
```

```python
# load in data, rename columns
data = pd.read_csv("../output/2024-05-17-14:27.predictions.tsv", sep="\t", header=None)
data.rename(columns={0:"receptor", 1:"ligand", 2:"prob"}, inplace=True)
data.sort_values(by="prob", ascending=False, inplace=True)
data.head(40)
```

```python
# gives best receptor match for each ligand 
best_match = data.loc[data.groupby('ligand')['prob'].idxmax()].reset_index(drop=True)
best_match.sort_values(by="prob", ascending=False)
```
